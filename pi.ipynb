```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nPoints = 10000
# https://docs.scipy.org/doc/numpy/reference/routines.random.html
# googled "numpy random", first hit
x = np.random.random_sample(nPoints)-0.5 
y = np.random.random_sample(nPoints)-0.5
```

```python
# most basic plotting functions
plt.scatter(x, y)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
plt.close()
```

```python
# numpy array arithmetic
r = np.sqrt(x**2 + y**2)
```

```python
# the np.where function is super useful
xIn = x[np.where(r < 0.5)]
yIn = y[np.where(r < 0.5)]
xOut = x[np.where(r > 0.5)]
yOut = y[np.where(r > 0.5)]
```

We want to get pi out of the area for a circle formula $$\pi r^2,$$ so pi is then equal to $\pi = \frac{1}{r^2}\frac{N_{in}}{N_{out}} A_{square} = \frac{1}{r^2}\frac{N_{in}}{N_{out}}$

```python
pi = float(len(xIn))/float(len(x))/0.5**2

# plotting two sets of data
plt.scatter(xIn, yIn, label = "points in r")
plt.scatter(xOut, yOut, label = "points out of r")
plt.plot(0,0, label = "$\pi = $"+str(pi), linewidth = 0)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc = 'upper right')
plt.show()
plt.close()
```
